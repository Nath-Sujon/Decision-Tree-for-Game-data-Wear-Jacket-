# Decision Tree for game data_Wear jacket decision

```python
import pandas as pd
df = pd.read_csv('game data.csv')
```

```python
df.head()
```

```python
df.drop('Days', axis = 'columns')
```

```python
df
```

```python
df.drop('Days', axis = 'columns', inplace = True)
```

```python
df.head()
```

# Lavel Encoding

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
df.Outlook = le.fit_transform(df['Outlook'])
df.Temprature = le.fit_transform(df['Temprature'])
df.Routine = le.fit_transform(df['Routine'])
```

```python
df.head()
```

```python
x = df.drop('Wear Jacket?', axis = 1)
```

```python
y = df['Wear Jacket?']
```

# Import Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()
```

```python
clf.fit(x,y)  #Train the ML model
```

```python
clf.predict([[1,0,0]])   #  Cold,Warm = (0,1);Indoor,Outdoor =(0,1); Sunny, Cloudy =(1,0); 
```

```python
clf.predict([[1,1,0]]) 
```

```python
clf.predict([[0,0,0]]) 
```
